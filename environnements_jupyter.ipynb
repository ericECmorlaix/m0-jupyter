## nbViewer :

Permet de partager l'affichage **statique** (non interactif) d'un notebook qui est disponible sur le web dans un dépot GitHub ou tout autre hébergement.

Il suffit de coller l'hyperlien du notebook à visualiser dans de la page : https://nbviewer.jupyter.org/

> Exemple des carnets de Yann GUELOU, collègue de math au lycée Notre Dame du Mur : http://sme.lycee.ecmorlaix.fr/YG/2019/python.html
>
> Remarque : le bouton <button class='fa fa-download icon-download btn btn-xs btn-default'></button> permet alors de télécharger une copie du fichier en local.
>

<a id="Binder"></a>
## Binder :

- https://mybinder.org/

Permet de partager une instance **dynamique** (interactive) mais temporaire d'un notebook qui est disponible sur le web depuis un dépot GitHub par exemple 

> **Attention**, il faut ajouter un [fichier de configuration](https://mybinder.readthedocs.io/en/latest/using.html#preparing-a-repository-for-binder) soit en [pip](https://github.com/binder-examples/requirements) soit en [conda](https://github.com/binder-examples/conda) pour indiquer à la machine virtuelle les paquets (Noyaux, Modules, Extensions) à installer.
